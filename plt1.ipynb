```python
x = [1,2,3,4,5]
y = np.array([1,4,9,16,24])
plt.plot(x,y)
plt.plot(x,y/2)
plt.plot(x,y/3)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
plt.show()
```

```python
x = [1,2,3,4,5]
y = np.array([1,4,9,16,24])
plt.plot(x,y, color='#41493F')
plt.plot(x,y/2, color='#464A84')
plt.plot(x,y/3, color='#00585A')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
plt.show()
```

```python
x = [1,2,3,4,5]
y = np.array([1,4,9,16,24])
plt.plot(x,y, color='#41493F', label='A')
plt.plot(x,y/2, color='#464A84', label='B')
plt.plot(x,y/3, color='#00585A', label='C')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5]
y = np.array([1,4,9,16,24])
plt.plot(x,y, color='#41493F', label='A')
plt.plot(x,y/2, color='#464A84', label='B')
plt.plot(x,y/3, color='#00585A', label='C')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.title('title')
plt.legend(fontsize=15)
plt.show()
```
