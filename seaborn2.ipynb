```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data
```

```python
sns.set(style='darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips_data,hue='time')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.relplot(x='total_bill',y='tip',data=tips_data,col='time',kind='scatter')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.relplot(x='total_bill',y='tip',data=tips_data,col='time',kind='scatter',height=3,aspect=2)
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.regplot(x='total_bill', y='tip', data=tips_data)
plt.show()
```

```python
sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data,hue='smoker')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips_data,hue='smoker', height=6, aspect=1.5)
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data)
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data,hue='sex')
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data,hue='sex',multiple='stack')
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data,hue='sex',multiple='dodge')
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set(style='darkgrid')
sns.histplot(x='day', data=tips_data,hue='sex',multiple='dodge', shrink=0.8)
plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.set_palette('Set1')
sns.histplot(x='day', data=tips_data,hue='sex',multiple='dodge', shrink=0.8)
plt.show()
```

```python
plt.figure(figsize=(9,6))


sns.histplot(x='day', data=tips_data,hue='sex',multiple='dodge', shrink=0.8,palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(9,6))

colors = ['coral', 'darkcyan']
sns.set_palette(colors)
sns.histplot(x='day', data=tips_data,hue='sex',multiple='dodge', shrink=0.8)
plt.show()
```

```python
plt.figure(figsize=(9,6))

colors = ['coral', 'darkcyan']
sns.histplot(x='day', data=tips_data,hue='sex',multiple='dodge', shrink=0.8,palette=colors)
plt.show()
```

```python
sns.set(style='darkgrid')
sns.pairplot(tips_data, hue='time', palette='Set1', diag_kind='hist')
plt.show()
```

```python
sns.set(style='darkgrid')
sns.pairplot(tips_data, hue='time', palette='Set1', diag_kind='hist',height=4, aspect=1)
plt.show()
```
