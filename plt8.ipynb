```python
from matplotlib import gridspec
```

```python
x = np.linspace(-np.pi, np.pi, 201)

y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize=(9,6))
grid = gridspec.GridSpec(nrows=2, ncols=1, height_ratios=[2,1])

ax1 = fig.add_subplot(grid[0])
ax1.plot(x, y_sin)

ax2 = fig.add_subplot(grid[1])
ax2.plot(x, y_cos)

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(9,6))

ax[0,0].plot(x, y_sin)
ax[0,1].plot(x ,y_cos)
ax[1,0].plot(x, y_sin)
ax[1,1].plot(x ,y_cos)

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(9,6))

ax[0,0].plot(x, y_sin)
ax[0,1].plot(x ,y_cos)
ax[1,0].plot(x, y_sin)
ax[1,1].plot(x ,y_cos)

plt.show()
```

```python
fig, ax = plt.subplots(4,1, figsize=(9,6))

ax[0].plot(x, y_sin)
ax[1].plot(x ,y_cos)
ax[2].plot(x, y_sin)
ax[3].plot(x ,y_cos)

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(9,6),sharex=True, sharey=True)

ax[0,0].plot(x, y_sin)
ax[0,1].plot(x ,y_cos)
ax[1,0].plot(x, y_sin)
ax[1,1].plot(x ,y_cos)

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(9,6),sharex=True, sharey=True)

ax[0,0].plot(x, y_sin)
ax[0,0].set(title='sin')

ax[0,1].plot(x ,y_cos)
ax[0,1].set(title='cos')

ax[1,0].plot(x, y_sin)
ax[1,1].plot(x ,y_cos)

plt.suptitle('sin&cos Graphs', fontsize=18)
plt.show()
```
